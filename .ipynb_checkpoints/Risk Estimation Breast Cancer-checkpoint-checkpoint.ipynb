# Train, Test, Split

Importing the train_test_split algorithm from sklearn to split the Training set

```python
from sklearn.model_selection import train_test_split
```

Before splitting the data, the target feature must be deleted from the dataframe then put in another variable, also the the target feature must be set aside then put into a seperate variable. In this instance we use the capital X and small y

Creating Independent AND Dependent features

```python
#X = Data_trainn.drop('cancer', axis = 1)
X = Data1.drop('cancer', axis = 1)
```

Importing the label Encoder to encode the target values then covert it to an array

```python
y = Data1['cancer']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

# Prediction (Random Forest Model)

Importing RandomForestClassifier from sklearn library

```python
from sklearn.ensemble import RandomForestClassifier
```

Imputing suitable parameters to achieve a high accuracy, then fiting data into the model (These parameters were gotten through hyperparameter tuning)

```python
#y_train.value_counts()
```

Over Sampling to handle imbalance

```python
class_weight = ({0:1,1:1000})
```

```python
import imblearn
from collections import Counter
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(0.78)
```

```python
X_train_os,y_train_os = os.fit_resample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
```

```python
RFC = RandomForestClassifier(class_weight =class_weight, n_estimators = 400,min_samples_split = 5,min_samples_leaf = 1,max_features = 'sqrt', max_depth = 780,criterion = 'gini', random_state = 100)
RFC.fit(X_train_os,y_train_os)
```

# Validation

Validation compares your trained prediction with your test data to confirm accuracy of the model also to ensure there is no form of overfitting or underfitting.

```python
y_pred_RFC = RFC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# View accuracy score
rf = round(accuracy_score(y_test, y_pred_RFC),2)*100
print(str(rf)+'%')

# View confusion matrix for test data and predictions
print(confusion_matrix(y_test, y_pred_RFC))

# View the classification report for test data and predictions
print(classification_report(y_test, y_pred_RFC))
```
